#### Please plot the impulse response function from a bivariate VAR(4) model with Taiwan's GDP and Stock Index data set.

```python
# import packages
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
%matplotlib inline

# read in data
data = pd.read_excel("Data.xlsx")

# transform "time" column to datatime object
data.time = pd.to_datetime(data.time)

# reset data's index
data.set_index("time", inplace = True)

data.head()
```

```python
# chose VAR model
VAR_4 = sm.tsa.VAR(data)

#fit data
VAR_4_result = VAR_4.fit(4)

#calculate impulse response function
VAR_4_result_irf = VAR_4_result.irf(60)

#plot impulse response function (orthogonalized, using the Cholesky decomposition)
VAR_4_result_irf.plot(orth = True)
```

**Reference:**


> **Vector Autoregressions tsa.vector_ar**
