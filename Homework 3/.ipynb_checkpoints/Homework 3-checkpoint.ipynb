#### Use the same realization of White noise $\varepsilon$'s (set $\sigma_{\varepsilon}^{2}$ = 1) to simulate and plot the following Gaussian process $Y_{t}$ (set $Y_{0}$ = E($Y_{t}$)) in a sample of size T = 150. Create new random data for each $Y_{t}$ every time before plotting:

(1) $Y_{t} = \varepsilon_{t}$

(2) $Y_{t} = \varepsilon_{t} +0.8 \varepsilon_{t-1}$

(3) $Y_{t} = \varepsilon_{t} -0.8 \varepsilon_{t-1}$

(4) $Y_{t} = 2 + \varepsilon_{t} +0.8 \varepsilon_{t-1}$

(5) $Y_{t} = \varepsilon_{t} +1.25 \varepsilon_{t-1}$

(6) $Y_{t} = 0.8Y_{t-1} + \varepsilon_{t}$

(7) $Y_{t} = -0.8Y_{t-1} + \varepsilon_{t}$

(8) $Y_{t} = 0.4 + 0.8Y_{t-1} + \varepsilon_{t}$

(9) $Y_{t} = 1.25Y_{t-1} + \varepsilon_{t}$

```python
# import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# define parameters
mean = 0
std = 1
sample_size = 151    #T
```

```python
# define function to create random data

def create_random_set(how_many_set_to_create):
    sample_set = pd.DataFrame(columns = range(how_many_set_to_create))
    
    for i in range(how_many_set_to_create):
        sample_set.iloc[:, i] = np.random.normal(mean, std, sample_size)
    
    return(sample_set)
```

### Plot (1) $Y_{t} = \varepsilon_{t}$ with three sample sets:

```python
sample_set = create_random_set(3)

plt.figure(figsize= (16, 7))
for i in range(3):
    plt.plot(sample_set.iloc[:, i], label = "sample_{}".format(i + 1), alpha = 0.8)

plt.xlabel("T")
plt.title(r"$Y_{t} = \varepsilon_{t}$", fontsize = 15)
plt.legend();
```

### Plot (2) $Y_{t} = \varepsilon_{t} +0.8 \varepsilon_{t-1}$ with three sample sets:

```python
# define function to calculate Yt, this function can be used to calculate Yt for Moving Average process
def calculate_Y_t_MA(data, constant, parameter):
    Y_t = data[1:].values + parameter*data[:-1].values + constant
    return(Y_t)
```

```python
sample_set = create_random_set(3)

plt.figure(figsize = (16, 7))
for i in range(3):
    plt.plot(calculate_Y_t_MA(sample_set.iloc[:, i], 0, 0.8), label = "sample_{}".format(i + 1, alpha = 0.8))

plt.xlabel("T")
plt.title(r"$Y_{t} = \varepsilon_{t} +0.8 \varepsilon_{t-1}$", fontsize = 15)
plt.legend();
```

### Plot (1) $Y_{t} = \varepsilon_{t}$ and (2)$Y_{t} = \varepsilon_{t} +0.8 \varepsilon_{t-1}$ with one sample set:

```python
plt.figure(figsize = (16, 7))

sample_set = create_random_set(1)
plt.plot(sample_set.drop(0, axis = 0).iloc[:, 0].values, label = r"$Y_{t} = \varepsilon_{t}$", alpha = 0.8)

sample_set = create_random_set(1)
plt.plot(calculate_Y_t_MA(sample_set.iloc[:, 0], 0, 0.8), label = r"$Y_{t} = \varepsilon_{t} +0.8 \varepsilon_{t-1}$", alpha = 0.8)

plt.xlabel("T")
plt.legend();
```

### Plot (2) $Y_{t} = \varepsilon_{t} +0.8 \varepsilon_{t-1}$,and (3) $Y_{t} = \varepsilon_{t} -0.8 \varepsilon_{t-1}$ with one sample set:

```python
plt.figure(figsize = (16, 7))

sample_set = create_random_set(1)
plt.plot(calculate_Y_t_MA(sample_set.iloc[:, 0], 0, 0.8), label = r"$Y_{t} = \varepsilon_{t} +0.8 \varepsilon_{t-1}$", alpha = 0.8)

sample_set = create_random_set(1)
plt.plot(calculate_Y_t_MA(sample_set.iloc[:, 0], 0, -0.8), label = r"$Y_{t} = \varepsilon_{t} -0.8 \varepsilon_{t-1}$", alpha = 0.8)

plt.xlabel("T")
plt.legend();
```

### Plot (2) $Y_{t} = \varepsilon_{t} +0.8 \varepsilon_{t-1}$,and (4) $Y_{t} = 2 + \varepsilon_{t} +0.8 \varepsilon_{t-1}$ with one sample set:

```python
plt.figure(figsize = (16, 7))

sample_set = create_random_set(1)
plt.plot(calculate_Y_t_MA(sample_set.iloc[:, 0], 0, 0.8), label = r"$Y_{t} = \varepsilon_{t} +0.8 \varepsilon_{t-1}$", alpha = 0.8)

sample_set = create_random_set(1)
plt.plot(calculate_Y_t_MA(sample_set.iloc[:, 0], 2, 0.8), label = r"$Y_{t} = 2 + \varepsilon_{t} +0.8 \varepsilon_{t-1}$", alpha = 0.8)

plt.xlabel("T")
plt.legend();
```

### Plot (2) $Y_{t} = \varepsilon_{t} +0.8 \varepsilon_{t-1}$,and (5) $Y_{t} = \varepsilon_{t} +1.25 \varepsilon_{t-1}$ with one sample set:

```python
plt.figure(figsize = (16, 7))

sample_set = create_random_set(1)
plt.plot(calculate_Y_t_MA(sample_set.iloc[:, 0], 0, 0.8), label = r"$Y_{t} = \varepsilon_{t} +0.8 \varepsilon_{t-1}$", alpha = 0.8)

sample_set = create_random_set(1)
plt.plot(calculate_Y_t_MA(sample_set.iloc[:, 0], 0, 1.25), label = r"$Y_{t} = \varepsilon_{t} +1.25 \varepsilon_{t-1}$", alpha = 0.8)

plt.xlabel("T")
plt.legend();
```

### Plot (6) $Y_{t} = 0.8Y_{t-1} + \varepsilon_{t}$, (7) $Y_{t} = -0.8Y_{t-1} + \varepsilon_{t}$,and (8) $Y_{t} = 0.4 + 0.8Y_{t-1} + \varepsilon_{t}$ with one sample set:

```python
# Define function to calculate Yt, this function can be used to calculate Yt for Autoregressive process
def calculate_Yt_AR(data, constant, parameter):
    temp = np.zeros(sample_size)
    for i in range(1,(sample_size)):
        if(i == 1):
            temp[i] = parameter * 0 + data[i] + constant
        else:
            temp[i] = parameter * temp[i - 1] + data[i] + constant
    return(temp[1:])
```

```python
plt.figure(figsize = (16, 7))

sample_set = create_random_set(1)
plt.plot(calculate_Yt_AR(sample_set.iloc[:, 0], 0, 0.8), label = r"$Y_{t} = 0.8Y_{t-1} + \varepsilon_{t}$")

sample_set = create_random_set(1)
plt.plot(calculate_Yt_AR(sample_set.iloc[:, 0], 0, -0.8), label = r"$Y_{t} = -0.8Y_{t-1} + \varepsilon_{t}$")

sample_set = create_random_set(1)
plt.plot(calculate_Yt_AR(sample_set.iloc[:, 0], 0.4, 0.8), label = r"$Y_{t} = 0.4 + 0.8Y_{t-1} + \varepsilon_{t}$")

plt.xlabel("T")
plt.legend();
```

### Plot (9) $Y_{t} = 1.25Y_{t-1} + \varepsilon_{t}$ with one sample set:

```python
plt.figure(figsize = (16, 7))

sample_set = create_random_set(1)
plt.plot(calculate_Yt_AR(sample_set.iloc[:, 0], 0, 1.25), label = r"$Y_{t} = 1.25Y_{t-1} + \varepsilon_{t}$")

plt.xlabel("T")
plt.legend();
```
