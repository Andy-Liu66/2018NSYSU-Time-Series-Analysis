#### Build up a stochastic model to the data set teacher gave from Box-Jenkins procedure, and predict values for five periods.
> Process:

> 1. Detrend (no need in this homework, because data had been preprocessed)
> 2. Configure parameters of ARMA model
> 3. Estimation
> 4. Diagnostic Checking
> 5. Prediction

```python
# import packages
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# read in data
data = pd.read_excel("Data.xlsx")

# create dummy dates for the data, because the model we are going to use require "dates" index
dates = pd.date_range('2017-1-1', periods=len(data), freq='D') #assume this is a daily data
# add the dates and the data to a new dataframe
data = pd.DataFrame({'dates': dates, 'data': data["values"]})
# set the dataframe index to be the dates column
data = data.set_index('dates')

# plot data
plt.style.use('seaborn')
plt.figure(figsize = (20, 5))
plt.plot(data);
```

### 2. Configure parameters of ARMA model
* Check plot of ACF, and PACF

```python
fig, ax = plt.subplots(2, 1, figsize = (20, 10))

ax1 = ax[0]
sm.graphics.tsa.plot_acf(data, ax = ax1)

ax2 = ax[1]
sm.graphics.tsa.plot_pacf(data, ax = ax2)

plt.tight_layout()
```

### 2. Configure parameters of ARMA model & 3. Estimation
> Actually, these two processes should be done at the same time, because from the ACF, PACF plot above, we can find that we can't use single AR or MA model in this data. As a result, we are going to chose the model that has the smallest AIC after estimatation model parameters: ARMA(p,q).

```python
# using nested for loop to create model for different combination of p, q(1:30 x 1:30)
# at the same time, calculate the AIC for each model, so we can chose the model that has the smallest AIC
table = pd.DataFrame()
for p in np.arange(30):
    for q in np.arange(30):
        try:
            model = sm.tsa.ARMA(data, (p,q)).fit(maxiter = 1000)  #using MLE by default
            temp = pd.DataFrame([p, q, model.aic])
            table = pd.concat([table, temp], axis = 1)
        except:
            pass

table = table.T
table.columns = ["p", "q", "AIC"]
table.reset_index(drop = True, inplace = True)
```

### 4. Diagnostic Checking

```python
# build ARMA(5, 2) model
ARMA_5_2 = sm.tsa.ARMA(data, (5, 2)).fit(maxiter = 1000)
```

Statsmodels package provide two statistic Ljung-Box and Box-Pierce. According to Statsmodels package's document it declare:

*"Ljung-Box and Box-Pierce statistic differ in their scaling of the autocorrelation function. Ljung-Box test is reported to have better small sample properties."*

I am not sure whether the data in this homework belongs to small sample or not. As a result, I am going to check both Ljung-Box and Box-Pierce statistic.

If the p-values of the test is not below 0.05, then we can not reject the null of no auto-correlation between the series and each of it's first n lags with > 95% confidence level. I am not sure how many lags should I check, so I am going to check all of it(199 lags).

```python
# First, p-values from Ljung-Box test
sm.stats.diagnostic.acorr_ljungbox(ARMA_5_2.resid, boxpierce = True, lags = 199)[1]
```

From the values above, we can find that almost all p-values are bigger than 0.05, except the p-values start from the 189th lags. I think the result is still acceptable, because we don't actually need to check so many lags as I did above(depends on the frequency of the data, I assume the data is a daily frequency, so at least it consider a semi-year pattern).

```python
# Sceond, p-values from Box-Pierce test
sm.stats.diagnostic.acorr_ljungbox(ARMA_5_2.resid, boxpierce = True, lags = 199)[3]
```

From the values above, we can find that all p-values are bigger than 0.05, so the result is acceptable.
As a result, we can not reject the null of no auto-correlation between the series and each of it's first 199 lags with > 95% confidence level.The ARMA(5, 2) model pass both Ljung-Box and Box-Pierce test.

```python
# we can also check ACF, PACF of the residual in ARMA(5, 2)
fig, ax = plt.subplots(2, 1, figsize = (20, 10))

ax1 = ax[0]
sm.graphics.tsa.plot_acf(ARMA_5_2.resid, ax = ax1)

ax2 = ax[1]
sm.graphics.tsa.plot_pacf(ARMA_5_2.resid, ax = ax2)

plt.tight_layout()
```

From the plot above, we can find that both auticorrelation and parital autocorrelation's values drop close to zero at lag(1) and above. However, we need to know that Ljung-Box and Box-Pierce are the test on the cumulative sum of autocorrelations, so it's different from just looking at individual autocorrelation in the plots. e.g. There could be many individually insignificant autocorrelation coefficients but they are all positive so they add up. As a result, we still need to check both Ljung-Box and Box-Pierce test.
